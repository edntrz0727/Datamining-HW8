```python
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
music_data = pd.read_csv('data.csv')
music_data
```

```python
music = music_data[['tempo','beats','label']]
music
```

```python
from sklearn.utils import resample
df_label = resample(music,n_samples=20)
# sampling
df_label
#keep the label data to check the clustering result with original dataset
```

```python
df = df_label[['tempo','beats']]
kmeans = KMeans(n_clusters=10, random_state=0,init='k-means++', n_init="auto").fit(df.values)
#sklearn only fitted with values
kmeans.labels_
#memory leak warning -> environment variable setting
```

```python
import random

tmp_max = df['tempo'].max()
tmp_min = df['tempo'].min()

beat_max = df['beats'].max()
beat_min = df['beats'].min()

x1 = random.uniform(tmp_min,tmp_max)
x2 = random.uniform(tmp_min,tmp_max)
y1 = random.randint(beat_min,beat_max)
y2 = random.randint(beat_min,beat_max)

arr = [x1,y1,x2,y2]
print('arr:',arr)
```

```python
kmeans.predict([[x1,y1],[x2,y2]])
```

```python
from sklearn.cluster import AgglomerativeClustering
clustering = AgglomerativeClustering(n_clusters=10,linkage='complete',metric='l2').fit(df.values)
clustering
```

```python
clustering.labels_
```

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=3,min_samples=1).fit(df.values)
dbscan
```

```python
dbscan.labels_
```
